# Digging into coaching goals

It's quite easy to look at the current set of goals and their status and jump to the conclusion that most goals go left unfinished. But what if we looked deeper at the goals? Could we predict the final status of a goal by just the type of goal that is created? Let's find out... 

# Data

There are few different tables used to dig into the current goals. All the tables live in the staging dataset. The main tables used were salesforceprod and d_programs. 

# Data Cleaning

Data cleaning took the longest due to the amount of inconsitent labeling, undocumented changes, and the ability for past and current employees to change fields within Salesforce. 

Problem 1: Inconsistent labeling. With the continued change of the company, the evolution of coaching and how we reach our members, labels have been changes and the usage of fields vary from person to person and from team to team. The first step was to identifty of certain labels could be combined as one. Paths and webinars became group coaching, onsite coaching rolled into one on one coaching, digital and amp became amp, etc. This helped limit the amount of labels. 

Problem 2: Lack of fields usage. Rows with multiple missing values were dropped. After dropping a good amount of fields, I took a step back and re-evalued what information I needed. To do this, I went directly to the source, Salesforce, and looked at required fields for all goals: goal status, modality, program name, type of goal, and created date.
